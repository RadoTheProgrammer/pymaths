```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use("grayscale")
```

```python
def serie(u,n,start=0):
    return np.sum([u(k) for k in range(start,n)])
```

```python
N = 40
X = list(range(N))
Y = [serie(lambda k:(1/2)**k,n) for n in X]
Y
```

```python
plt.plot(X,Y, label="Somme des (1/2)**k")
plt.legend()
```

```python
N = 1000
X = [n for n in range(1,N)]
Y = [serie(lambda k:(1/k),n,1) for n in X]
```

```python
plt.plot(X,Y, label="Somme des 1/k")
plt.legend()
```
