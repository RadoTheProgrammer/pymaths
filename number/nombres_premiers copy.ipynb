# définir sieves


```python
def is_multiple(a,b):
    return a != b and a%b == 0
```

```python
is_multiple(12,3)
#> True
```

```python
is_multiple(13,3)
```

```python
#crible d'Eratosthène
def sieves(n):
    primes = list(range(2,n+1))
    idx = 0
    for p in primes:
        primes = list(filter(lambda x:not is_multiple(x,p), primes))
    return primes
```

```python
sieves(100)
```

```python
len(sieves(500))
```

```python
len(sieves(6000))
```

```python
from math import log
import matplotlib.pyplot as plt
```

```python
max = 10000
ns = [n for n in range(10,max,10)]
ns
```

```python
primes = sieves(max)
```

```python
pi = {n:len([x for x in primes if x <= n]) for n in ns}
pi
```

```python
freq = {n:pi[n] / n for n in ns}
freq
```

```python
chck = {n:freq[n] * log(n) for i,n in enumerate(ns)}
```

```python
plt.plot(ns,chck.values())
```
