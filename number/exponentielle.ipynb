```python
def compound_interest(r,n,d):
    """
    r le taux d'interêt
    n le nombre de fois que l'intérêt est composé par an
    d le nombre d'années
    """
    return (1 + r/n)**(n*d)
```

```python
compound_interest(0.05,1,1)
```

```python
compound_interest(0.05,12,1)
```

```python
compound_interest(1.0,1,1)
```

```python
def exp_euler(r,max_iter=20):
    fact = 1
    res = 0.0
    r_n = 1
    for n in range(max_iter):
        fact = fact * 1 if n == 0 else fact * n
        res += r_n / fact
        r_n = r * r_n # r_n = r**n
    return res
```

```python
import matplotlib.pyplot as plt
plt.plot(range(1,20),[exp_euler(n) for n in range(1,20)])
```

```python
l=range(1,1001)
plt.plot(l, [exp_euler(1, n) for n in l], label='exp_euler')
plt.plot(l, [compound_interest(1, n, 1) for n in l], label='compound')
plt.legend()
```
